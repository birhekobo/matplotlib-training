## 1. Why Environment Setup Matters

A **reproducible environment** ensures:
- Your code works on another machine
- Package versions are consistent
- You avoid "works on my machine" problems
- Collaboration and deployment become straightforward

### Two popular approaches:

| Tool | Pros | Cons |
|------|------|------|
| **venv** | Built-in, lightweight | Python-only packages |
| **conda** | Cross-language, handles non-Python deps (netCDF, geospatial libs) | Heavier, more disk space |

We will use **conda** for this course because cartopy and netCDF4 have non-Python dependencies.

## 4. Verify Installation

Run the cell below to confirm all packages are installed and the CHIRPS data loads correctly.

````python
# Check package versions
import sys
print(f'Python: {sys.version}')

import numpy as np
print(f'numpy: {np.__version__}')

import pandas as pd
print(f'pandas: {pd.__version__}')

import matplotlib as mpl
print(f'matplotlib: {mpl.__version__}')

import xarray as xr
print(f'xarray: {xr.__version__}')

import cartopy
print(f'cartopy: {cartopy.__version__}')

print('\nAll packages installed successfully!')
````

````python
# Load CHIRPS data
ds = xr.open_dataset('../chirps_1981_2022.nc')
print(f'Dataset dimensions: {dict(ds.dims)}')
print(f'Time range: {ds.time.values[0]} to {ds.time.values[-1]}')
print(f'Latitude range: {float(ds.latitude.min().values):.3f} to {float(ds.latitude.max().values):.3f}')
print(f'Longitude range: {float(ds.longitude.min().values):.3f} to {float(ds.longitude.max().values):.3f}')
print(f'Precip data type: {ds.precip.dtype}')
print(f'Precip units: mm/month')
````

### Basic plot to verify everything works end-to-end

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Extract a single time step
precip_jan2022 = ds.precip.sel(time='2022-01-01')

fig, ax = plt.subplots(figsize=(9, 7),
                       subplot_kw={'projection': ccrs.PlateCarree()})

im = ax.pcolormesh(ds.longitude, ds.latitude, precip_jan2022.values,
                   cmap='Blues', shading='auto')
ax.coastlines(resolution='50m', linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linewidth=0.5, alpha=0.6)
ax.add_feature(cfeature.LAKES, color='lightblue', alpha=0.3)
gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False,
                  linewidth=0.3, color='gray', alpha=0.5)
gl.top_labels = False
gl.right_labels = False
ax.set_title('CHIRPS Rainfall — January 2022\nEthiopia', fontsize=13, fontweight='bold')
cbar = fig.colorbar(im, ax=ax, label='mm / month', shrink=0.7)

plt.tight_layout()
plt.show()
print('End-to-end pipeline: DATA LOADED + PLOT RENDERED ✓')
````

Version control is essential for tracking changes and collaborating.

## Mini-Project: Setup Verification Dashboard

Create a notebook cell that:
- Prints all installed package versions
- Loads the CHIRPS dataset
- Computes and prints the mean, min, and max precipitation
- Creates a 2-panel figure showing:
  1. The spatial map of **mean** rainfall (climatology)
  2. A **histogram** of all pixel values at a specific time
- Saves the figure as `setup_verification.png`

````python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

ds = xr.open_dataset('../chirps_1981_2022.nc')
clim = ds.precip.mean(dim='time')

print(f'Mean: {float(clim.mean().values):.1f} mm/month')
print(f'Min:  {float(clim.min().values):.1f} mm/month')
print(f'Max:  {float(clim.max().values):.1f} mm/month')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5),
                               subplot_kw={'projection': ccrs.PlateCarree()})

# Panel 1: spatial map
im = ax1.pcolormesh(ds.longitude, ds.latitude, clim.values,
                    cmap='viridis', shading='auto')
ax1.coastlines(resolution='50m', linewidth=0.8)
ax1.add_feature(cfeature.BORDERS, linewidth=0.5, alpha=0.6)
ax1.set_title('Mean Monthly Rainfall (Climatology)')
fig.colorbar(im, ax=ax1, label='mm/month', shrink=0.7)

# Panel 2: histogram
ax2 = ax2  # no projection for histogram
_ = ax2.hist(clim.values.ravel(), bins=80, color='steelblue', edgecolor='white', linewidth=0.3)
ax2.set_title('Distribution of Mean Rainfall')
ax2.set_xlabel('mm / month')
ax2.set_ylabel('Pixel count')
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('setup_verification.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved as setup_verification.png ✓')
````

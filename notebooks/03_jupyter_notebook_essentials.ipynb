## 1. Notebook Interface Overview

A Jupyter notebook is a **JSON document** (`.ipynb`) containing an ordered list of **cells**. There are three cell types:

| Cell Type | Purpose |
|-----------|---------|
| **Code** | Executable Python (or other language) code |
| **Markdown** | Formatted text, images, equations, tables |
| **Raw** | Unformatted text (not rendered) |

### Two modes:
- **Edit mode** (green border) — type into the cell
- **Command mode** (blue border) — keyboard shortcuts operate on cells

Press `Esc` to enter command mode, `Enter` to enter edit mode.

## 3. Code Cells and Execution

Code cells contain executable Python. Each cell has an **execution counter** `In [n]` that increments with each run.

### Execution order matters!
- Cells can be run in **any order** — the notebook remembers the state
- Always run cells in sequence from top to bottom
- A common source of bugs: running cells out of order (stale variables)
- Use **Kernel → Restart & Run All** to get a clean, sequential execution

### Cell output:
- **Last expression** is automatically displayed (no `print()` needed for simple values)
- Use `print()` for multiple outputs or formatted strings
- Matplotlib figures appear inline with `%matplotlib inline`

````python
# Code cell example — the last value is printed automatically
import numpy as np
data = np.random.randn(100)
data.mean()
````

````python
# Use print for multiple explicit outputs
print(f'Mean: {data.mean():.3f}')
print(f'Std:  {data.std():.3f}')
````

## 5. Magic Commands

Magic commands (or "magics") are special commands prefixed with `%` (line magic) or `%%` (cell magic). They extend Jupyter's capabilities.

| Magic | Purpose |
|-------|---------|
| `%matplotlib inline` | Show plots inline in the notebook |
| `%matplotlib notebook` | Interactive plots (zoom, pan) |
| `%timeit` | Time the execution of a statement (average of runs) |
| `%time` | Time a single execution |
| `%load` | Load code from a file or URL into a cell |
| `%who` | List all variables in the namespace |
| `%whos` | List variables with type and size |
| `%env` | List/set environment variables |
| `%run` | Run an external Python script |
| `%debug` | Enter post-mortem debugger after an error |
| `%pdb` | Auto-enable debugger on exception |
| `%lsmagic` | List all available magics |
| `%%capture` | Capture stdout/stderr of a cell |
| `%%writefile` | Write cell contents to a file |
| `%%html` | Render cell as HTML |
| `%%bash` | Run cell as bash command |

````python
# List all available magics
%lsmagic
````

````python
# %timeit — time expression execution
import numpy as np
%timeit np.random.randn(1000).sum()
````

````python
# %who / %whos — inspect variables
a = 42
b = 'hello'
c = np.array([1, 2, 3])
%whos
````

````python
# %%writefile — save cell content to a Python file
%%writefile hello_chirps.py
print('Hello from CHIRPS! This file was created by a Jupyter magic.')
print('CHIRPS covers Ethiopia at 0.05° resolution.')
````

````python
# %run — execute the saved script
%run hello_chirps.py
````

## 6. Interactive Exploration of CHIRPS Data

Let us combine everything to explore the dataset interactively.

````python
%matplotlib inline
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

ds = xr.open_dataset('../chirps_1981_2022.nc')
ds
````

````python
# Examine a specific location
lat = 11.0  # near Bahir Dar
lon = 37.5

ts = ds.precip.sel(latitude=lat, longitude=lon, method='nearest')
print(f'Location: lat={float(ts.latitude.values):.3f}, lon={float(ts.longitude.values):.3f}')
print(f'Time series length: {len(ts)} months')
print(f'Annual mean: {float(ts.groupby("time.year").mean().mean().values):.1f} mm/month')
````

````python
# Quick interactive plot
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(ts.time.values, ts.values, color='steelblue', linewidth=0.8)
ax.set_title(f'Rainfall at lat={lat}, lon={lon} — CHIRPS 1981–2022')
ax.set_xlabel('Time')
ax.set_ylabel('Precipitation (mm/month)')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Use %timeit on a computation
%timeit ds.precip.mean(dim='time').values
````

````python
# Environment variables with %env
%env OMP_NUM_THREADS 4
import os
print(f'OMP_NUM_THREADS = {os.environ.get("OMP_NUM_THREADS", "not set")}')
````

## 7. Exporting Notebooks

Notebooks can be exported to various formats:

### Via the UI:
**File → Download As →**
- `Python (.py)` — pure script, cell structure lost
- `HTML (.html)` — standalone web page
- `Markdown (.md)` — markdown + code blocks
- `PDF via LaTeX (.pdf)` — requires LaTeX installation
- `ReStructured Text (.rst)` — for documentation
- `Slides (.slides.html)` — reveal.js presentation

### Via command line:
```bash
# Convert to HTML
jupyter nbconvert --to html notebook.ipynb

# Convert to Python script
jupyter nbconvert --to script notebook.ipynb

# Execute and convert to HTML
jupyter nbconvert --execute --to html notebook.ipynb

# Convert to slides
jupyter nbconvert --to slides notebook.ipynb --post serve
```

## Mini-Project: Interactive Notebook Report

Create a notebook that serves as an **interactive report** on the CHIRPS dataset. It must include:

1. A markdown title and introduction describing the dataset
2. A table of dataset dimensions and key attributes
3. Code that loads and displays the dataset
4. A multi-panel figure (2×2) showing:
   - Monthly climatology map
   - Time series at a point in the Ethiopian highlands
   - Histogram of precipitation values (all time, all pixels)
   - A seasonal cycle (12-month climatology at that point)
5. Use `%timeit` to show performance of the mean computation
6. Export the notebook to HTML

````python
# Mini-project: multi-panel figure
fig, axes = plt.subplots(2, 2, figsize=(14, 10),
                         subplot_kw={'projection': ccrs.PlateCarree()})

clim = ds.precip.mean(dim='time')
ts_point = ds.precip.sel(latitude=11.5, longitude=39.5, method='nearest')
monthly_clim = ts_point.groupby('time.month').mean()

# (a) Climatology map
im = axes[0, 0].pcolormesh(ds.longitude, ds.latitude, clim.values,
                            cmap='viridis', shading='auto')
axes[0, 0].coastlines(resolution='50m', linewidth=0.8)
axes[0, 0].add_feature(cfeature.BORDERS, linewidth=0.5, alpha=0.6)
axes[0, 0].set_title('(a) Mean Monthly Rainfall')
fig.colorbar(im, ax=axes[0, 0], label='mm/month', shrink=0.6)

# (b) Time series
ax_ts = axes[0, 1]
ax_ts.remove()
ax_ts = fig.add_subplot(2, 2, 2)
ax_ts.plot(ts_point.time.values, ts_point.values, color='steelblue', linewidth=0.6)
ax_ts.set_title('(b) Time Series at (11.5°N, 39.5°E)')
ax_ts.set_xlabel('Time')
ax_ts.set_ylabel('mm/month')
ax_ts.grid(alpha=0.3)

# (c) Histogram
ax_hist = axes[1, 0]
ax_hist.remove()
ax_hist = fig.add_subplot(2, 2, 3)
_ = ax_hist.hist(clim.values.ravel(), bins=100,
                  color='steelblue', edgecolor='white', linewidth=0.3)
ax_hist.set_title('(c) Distribution of Mean Rainfall')
ax_hist.set_xlabel('mm/month')
ax_hist.set_ylabel('Pixel count')
ax_hist.grid(alpha=0.3)

# (d) Seasonal cycle
ax_seas = axes[1, 1]
ax_seas.remove()
ax_seas = fig.add_subplot(2, 2, 4)
ax_seas.plot(monthly_clim.month.values, monthly_clim.values,
             marker='o', color='darkorange', linewidth=2)
ax_seas.set_title('(d) Seasonal Cycle at (11.5°N, 39.5°E)')
ax_seas.set_xlabel('Month')
ax_seas.set_ylabel('mm/month')
ax_seas.set_xticks(range(1, 13))
ax_seas.grid(alpha=0.3)

plt.suptitle('CHIRPS Ethiopia Interactive Report', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

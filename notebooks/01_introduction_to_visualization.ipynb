## 1. What is Data Visualization?

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

### Why it matters:
- **Human perception** — our brains process visual information 60 000× faster than text
- **Pattern recognition** — trends, clusters, and anomalies leap out visually
- **Communication** — a well-designed chart tells a story that tables cannot
- **Exploration** — interactive graphics let us ask & answer questions quickly

> *"The greatest value of a picture is when it forces us to notice what we never expected to see."* — John Tukey

---
## 2. Principles of Effective Graphics

### Edward Tufte's Principles

| Principle | Meaning |
|-----------|---------|
| **Data-ink ratio** | Maximise ink that represents data, minimise non-data ink |
| **Chartjunk** | Avoid unnecessary decorations (3D, excessive gridlines, clip art) |
| **Lie factor** | The size of effect shown should be proportional to the size of effect in data |
| **Small multiples** | Show complex data across multiple small panels |

### Alberto Cairo's Principles

1. **Truthful** — represent data accurately
2. **Functional** — make patterns easy to see
3. **Beautiful** — aesthetically pleasing (but not at cost of clarity)
4. **Insightful** — reveal something unexpected
5. **Enlightening** — increase understanding

### The Data-Ink Ratio

$$\text{Data-Ink Ratio} = \frac{\text{Ink used for data}}{\text{Total ink used to print the graphic}}$$

**Goal:** Remove elements that do not convey information (gridlines, borders, unnecessary labels).

## 3. Types of Charts and When to Use Them

| Chart Type | Best For | Example Use |
|------------|----------|-------------|
| **Line chart** | Trends over time | Monthly rainfall 1981–2022 |
| **Bar chart** | Comparing categories | Mean rainfall by month |
| **Histogram** | Distribution of values | Distribution of pixel rainfall |
| **Scatter plot** | Relationship between 2 variables | Latitude vs rainfall |
| **Box plot** | Spread and outliers | Rainfall by season |
| **Heatmap / pcolormesh** | Spatial data | Map of Ethiopia rainfall |
| **Violin plot** | Distribution + density | Rainfall by decade |

---
## 4. Common Mistakes & Bad Practices

- ❌ **Truncated y-axis** — exaggerates differences
- ❌ **3D charts** — distort perception, reduce readability
- ❌ **Pie charts with many slices** — hard to compare angles
- ❌ **Dual y-axes** — misleading correlations
- ❌ **Rainbow colour maps** — not perceptually uniform
- ❌ **Overplotting** — too many points hide patterns
- ❌ **Missing labels / units** — chart is meaningless
- ❌ **Cherry-picked time range** — manipulates narrative

### Pre-attentive Attributes
Our visual system processes certain properties instantly:
- Position, Size, Shape, Colour (hue & intensity), Orientation, Motion

## 5. First Steps: Load CHIRPS Data & Plot

We will use `xarray` to load the NetCDF file and `matplotlib` to create a simple map.

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load CHIRPS data
ds = xr.open_dataset('../chirps_1981_2022.nc')
print(ds)
print(f'Shape of precip: {ds.precip.shape}')
```

```python
# Compute climatology (mean over all months)
clim = ds.precip.mean(dim='time')
print(f'Mean rainfall over Ethiopia (mm/month): {float(clim.mean().values):.1f}')
```

```python
# Map of mean annual rainfall
fig, ax = plt.subplots(figsize=(10, 8),
                       subplot_kw={'projection': ccrs.PlateCarree()})

im = ax.pcolormesh(ds.longitude, ds.latitude, clim.values,
                   cmap='viridis', shading='auto')

ax.coastlines(resolution='50m', linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linewidth=0.5, alpha=0.7)
ax.add_feature(cfeature.LAKES, color='lightblue', alpha=0.4)
ax.set_title('Mean Monthly Rainfall — Ethiopia\nCHIRPS 1981–2022',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
cbar = fig.colorbar(im, ax=ax, label='mm / month', shrink=0.7)

plt.tight_layout()
plt.show()
```

### Discussion

- What patterns do you see? (Highlands vs lowlands, seasonal zones)
- Is the colour map appropriate? (Viridis is perceptually uniform)
- Could we improve the data-ink ratio?

## 6. Critique a Chart

Let us intentionally create a **bad** chart to see what not to do.

```python
# A deliberately bad chart — truncated y-axis, too many colours, missing labels
point_data = ds.precip.isel(latitude=100, longitude=300).values

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(point_data[:60], color='red', linewidth=4)
ax.set_ylim(50, 120)  # truncated y-axis
ax.set_title('Rainfall')  # no units, no x-label
plt.show()
```

**What is wrong?**
1. Truncated y-axis exaggerates variation
2. Missing x-label (time? months?)
3. Title is vague — no units, no location
4. Single colour — no differentiation
5. No grid or reference lines
